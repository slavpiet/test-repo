# is_latin

```python
is_latin = data_csv_preprocessed['passwords'].replace(regex='[^\x00-\x7F]+', value = 0)
is_latin[is_latin !=0] = 1

data_csv_preprocessed['is_latin'] = is_latin
data_csv_preprocessed.sample(20)
```

# is_email

```python
is_email = data_csv_preprocessed['passwords'].replace(regex='^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$', value = 1)
is_email[is_email != 1] = 0
is_email.sample(20)

data_csv_preprocessed['is_email'] = is_email
data_csv_preprocessed.sample(20)
```

```python
def compute_distribution_for_places(password, dict_compute):
  for idx, sign in enumerate(password):
    try: 
      dict_compute[idx][sign] += 1
    except KeyError:
      dict_compute[idx][sign] = 1
  return dict_compute

def analyze_password_place( df: pd.DataFrame, max_lenght:int) -> dict:
  dict_comput = {idx: {} for idx in range(max_lenght)}
  for row in df.itertuples(index=False):
    dict_comput = compute_distribution_for_places(row.passwords, dict_comput)

  return dict_comput
```
